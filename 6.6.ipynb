```python
x = data.iloc[:,0:1].values
y = data.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 1000)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
sal = Sequential()
```

```python
sal.add(Dense(1,kernel_initializer="random_uniform",activation="relu"))
```

```python
sal.add(Dense(2,kernel_initializer="random_uniform",activation="relu"))
```

```python
sal.add(Dense(1,kernel_initializer="random_uniform",activation="relu"))
```

```python
sal.compile("adam","mean_squared_error",metrics = ["MSE"])
```

```python
sal.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
ypred = sal.predict(x_test)
ypred
```
